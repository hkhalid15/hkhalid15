```python
# Import Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Import ride data into pandas from CSV

ride_data_df = pd.read_csv("ride_data.csv")
ride_data_df.head()
```

```python
# Import city data into pandas from CSV

city_data_df = pd.read_csv("city_data.csv")
city_data_df.head()
```

```python
tot_drivers_city = city_data_df["driver_count"]
tot_drivers_city
```

```python
city_group = ride_data_df.groupby('city')

avg_fares = city_group['fare'].mean()

avg_fares
```

```python
tot_rides_city = city_group['ride_id'].count()
tot_rides_city
```

```python
plt.title("Pyber RideSharing Data (2016)")
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average Fare ($)")

plt.scatter(tot_rides_city, avg_fares, s=tot_drivers_city)
plt.show()
```

```python
merged_df = pd.merge(ride_data_df, city_data_df, on="city")
merged_df
```

```python
citygrp = city_data_df.groupby('type')

tot_drivers = citygrp['driver_count'].sum()

tot_drivers
```

```python
drivers_chart = tot_drivers.plot(kind='pie', figsize=(9,9))

# Set the xlabel and ylabel using class methods
drivers_chart.set_xlabel("City Type")
drivers_chart.set_ylabel("Number of Drivers")


plt.show()
```
